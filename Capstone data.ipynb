## Checking Missing Values

```python
#Lets make a function for number of missing values and percentage of missing values
def null_values(x):
    sum_total=x.isnull().sum().sort_values(ascending=False)
    percentage=((sum_total/len(x))*100).sort_values(ascending=False)
    
    return pd.concat([sum_total,percentage],axis=1,keys=['Sum_Total','Percentage_Missing'])
```

## Merging Data

## Merging Bureau and Bureau_Balance data

```python
bureau_balance.info()

# 1 categorical and 2 numerical columns
```

```python
# bureau balance and bureau data
#numerical variables
col=bureau_balance.groupby('SK_ID_BUREAU').mean().reset_index()

#categorical columns
bureau_balance_cat=bureau_balance.select_dtypes('object')
bureau_balance_cat['SK_ID_BUREAU']=bureau_balance['SK_ID_BUREAU']
# using mode and resetting the index
bureau_balance_cat=bureau_balance_cat.groupby('SK_ID_BUREAU').agg(lambda x:stats.mode(x)[0]).reset_index()

#merge numerical and categorical
col=col.merge(bureau_balance_cat,on='SK_ID_BUREAU',how='left')

#null values shows that there is no value present will replace with zero
#because in left join priority is always left

col.update(col[bureau_balance_cat.columns].fillna(0))

# merging with bureau 
Bureau=Bureau.merge(col,on='SK_ID_BUREAU',how='left')

#null values shows that there is no value present will replace with zero
#because in left join priority is always left

Bureau.update(Bureau[col.columns].fillna(0))
```

## Joining bureau and application_train

```python
Bureau.info()
# 4 categorical and 13 numerical columns
```

```python
#Bureau data numerical column
col1 = Bureau.drop(['SK_ID_BUREAU'], axis = 1).groupby(by=['SK_ID_CURR']).mean().reset_index()
# Renaming columns of new data frame prefixing bureau
col1.columns = ['BUREAU_'+column if column !='SK_ID_CURR' else column for column in col1.columns]
#merging in new 
application_bureau = Application_train.merge(col1, on='SK_ID_CURR', how='left')

#null values shows that there is no value present will replace with zero
#because in left join priority is always left

application_bureau.update(application_bureau[col1.columns].fillna(0))

# we have stored our data in new data frame with numerical column using mean of values

# Dealing categorical columns


#extract categorical columns
Bureau_cat=Bureau.select_dtypes('object')
#Bureau_cat=pd.get_dummies(Bureau.select_dtypes('object'))
#assign SK_ID_CURR to cat 
Bureau_cat['SK_ID_CURR']=Bureau['SK_ID_CURR']
#groupby using mode and reseting the index
Bureau_cat=Bureau_cat.groupby('SK_ID_CURR').agg(lambda x:stats.mode(x)[0]).reset_index()

#renaming columns and prefixing bureau infront of every col name except SK_ID_CURR so that we can understand source of varaible 
# and we can use this info while filling null values and treating outliers

Bureau_cat.columns=['BUREAU_'+column if column !='SK_ID_CURR' else column for column in Bureau_cat.columns]

#merging values with application train
application_bureau=application_bureau.merge(Bureau_cat,on='SK_ID_CURR',how='left')

#null values shows that there is no value present will replace with zero
#because in left join priority is always left

application_bureau.update(application_bureau[Bureau_cat.columns].fillna(0))

#Shape of final bureau and application train after merging both into application_bureau

print('Shape of application_bureau=',application_bureau.shape)
```

## Joining Previous_application data with application_bureau  data

```python
Previous_application.info()

# 15 are categorical and 11 are numerical columns
```

```python
#No. of past loans per customer
col2=Previous_application[['SK_ID_CURR','SK_ID_PREV']].groupby(by=['SK_ID_CURR'])['SK_ID_PREV'].count().reset_index().rename(columns={'SK_ID_PREV':'Previous_application_count'})

application_bureau_prev=application_bureau.merge(col2,how='left',on='SK_ID_CURR')

#null values shows that there is no value present will replace with zero
#because in left join priority is always left


application_bureau_prev.update(application_bureau_prev[col2.columns].fillna(0))

#Combining Numerical features
col3=Previous_application.drop('SK_ID_PREV',axis=1).groupby('SK_ID_CURR').mean().reset_index()

prev_columns = ['PREV_'+column if column != 'SK_ID_CURR' else column for column in col3.columns ]

#print(prev_columns)

col3.columns=prev_columns

application_bureau_prev=application_bureau_prev.merge(col3,on='SK_ID_CURR',how="left")

#null values shows that there is no value present will replace with zero
#because in left join priority is always left


application_bureau_prev.update(application_bureau_prev[col3.columns].fillna(0))

# Categorical columns
prev_categorical=previous_application.select_dtypes('object')
prev_categorical['SK_ID_CURR']=previous_application['SK_ID_CURR']

prev_categorical=prev_categorical.groupby('SK_ID_CURR').agg(lambda x:stats.mode(x)[0]).reset_index()

prev_categorical.columns=['PREV_'+column if column != 'SK_ID_CURR' else column for column in prev_categorical.columns ]

application_bureau_prev=application_bureau_prev.merge(prev_categorical,how='left',on='SK_ID_CURR')

#null values shows that there is no value present will replace with zero
#because in left join priority is always left

application_bureau_prev.update(application_bureau_prev[prev_categorical.columns].fillna(0))
```

## Outlier Capping  

```python
# In age and average workhours, there are values which could not be possible so we are capping those
# outliers alongwith other outliers so that we dont give our model wrong learnings.
```

```python
df=pd.concat([nums,cat],axis=1)
```

```python
for i in nums.columns:
    if i!='TARGET':
        q1=df[i].quantile(.25)
        q3=df[i].quantile(.75)
        iqr=q3-q1
        upper_limit=q3+1.5*iqr
        lower_limit=q1-1.5*iqr
        df.loc[df[i]>upper_limit,i]=upper_limit
        df.loc[df[i]<lower_limit,i]=lower_limit
```
